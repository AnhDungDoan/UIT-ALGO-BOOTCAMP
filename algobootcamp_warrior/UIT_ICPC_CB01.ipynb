## **Tóm tắt đề bài**: 
Cho bảng kí tự A = [a..z], kích thước MxN. Q truy vấn, mỗi truy vấn gồm một xâu S, cần kiểm tra xem S có tồn tại trong A không (có thể đọc từ trên xuống dưới hoặc từ trái qua phải)

**Ràng buộc:** 
  * $1<= M*N <=10^5$
  * $1<= Q <= 10^4$
  * $len(S[i])<=10$

## **Mô tả:**
Sử dụng Trie cho độ phức tạp tối ưu nhất \
Sau đây là code sử dụng Trie con trỏ động, thay vì khai báo mảng tĩnh (code submit trong đề bị **limited memory** nhưng đây là 1 trong những cách có thể sử dụng vì có nhiều lợi ích của nó, chi tiết tham khảo  [link](https://leduythuccs.github.io/2018-05-23-c-u-tr-c-d-li-u-trie/) này.

***Lỗi mắc phải:***


*   Truyền tham số char array vào hàm
*   Tính toán bộ nhớ ở TH xấu nhất phù hợp với đề bài

***Mở rộng*** \

```python
/*
	author: gnudnaod
	create: ..............
*/

#include <bits/stdc++.h>
#define F(i,a,b) for (int i = a; i <= b; i++)
#define _F(i,a,b) for (int i = a; i >= b; i--)
#define ll long long
#define push_back pb

using namespace std;

const int maxn = 100;

struct trieNode
{
    trieNode* child[27];
    int cnt = 0;
    trieNode()
    {
        F(i,0,26) child[i] = NULL;
    }
};

trieNode* root = new trieNode(); // root
int n, m, q;

void addNode(string tmp)
{
    int l = tmp.size();

    trieNode* p = root;
    F(i,0,l-1)
    {
        int nxt = tmp[i] - 'a';
        if (p->child[nxt] == NULL)
            p->child[nxt] = new trieNode();
        p = p->child[nxt];
    }
    p->cnt++;
}

bool findTrie(string s)
{
    int l = s.size();
    trieNode* p = root;
    F(i,0,l-1)
    {
        int nxt = s[i] - 'a';
        if (p->child[nxt] == NULL) return false;
            else
                p = p->child[nxt];
    }

    return true;
}

int main()
{
    cin >> n >> m >> q;
    char a[n+10][m+10];
    F(i,1,n) F(j,1,m) cin >> a[i][j];

    // add node into trie
    F(i,1,n)
    {
        F(j,1,m)
        {
            string tmp="";
            F(k,0,9)
                if (j+k <= m) tmp += a[i][j+k]; else break;
            addNode(tmp);
        }
    }

    F(j,1,m)
    {
        F(i,1,n)
        {
            string tmp="";

            F(k,0,9)
                if (i+k <= n) tmp += a[i+k][j]; else break;
            addNode(tmp);
        }

    }
    //

    F(i,1,q)
    {
        string s;
        cin >> s;
        if (findTrie(s)) cout << "1"; else cout << "0";
    }
    return 0;
}
```

# Bài 2: True Expression

## **Tóm tắt đề bài**: 
Cho N và S, với i = [1..n] thêm dấu "+" hoặc "-" trước mỗi i để được tổng = S, nếu không có in ra "NO_SOLUTION"

**Ràng buộc:**
- $1<=N<=100$
- $1<=S<=10000$

## **Mô tả:**
Kỹ thuật sử dụng:
*   QHĐ tổng cơ bản không truy vết
*   Toán học


Độ phức tạp bài toán: O(n * S) \
**Lưu ý**: Ưu tiên kết quả "NO_SOLUTION" trước khi xét các trường hợp tiếp theo.

***Mở rộng:***

https://vnoi.info/problems/BLOPER/ \
https://vnoi.info/problems/BLOPER2/



```python
arr = [int(i) for i in input().split()]
n, s = arr[0], arr[1]
ans = [' ' for i in range(n+10)]

if s > n*(n + 1)/2:
    print("NO_SOLUTION")
elif (s - (n*(n + 1)/2)) % 2 != 0:
    print("NO_SOLUTION")
else:
    for i in range(1, n+1):
        ans[i] = '+'
    sum = 0
    pres = ((n*(n + 1)/2) - s) /2
    for i in range(1, n+1):
        sum += i
        if sum > pres:
            for j in range(1, i+1):
                if j != sum - pres:
                    ans[j] = '-'
            break

    for i in range(1, n+1):
        print(ans[i], end = "")
```

# Bài 3: 	Top k Hits

## **Tóm tắt đề bài**: 
Cho dãy N phần tử ban đầu đều bằng 0, Q truy vấn có dạng, mỗi truy vấn bắt đầu bằng x. Có 2 loại truy vấn, 
 - x = 1 là tăng đoạn [l,r] lên v đơn vị
 - x = 2 là tìm k phần tử lớn nhất trong dãy

**Ràng buộc:**
- $1<=N,Q<=50000$
- $1<=l,r<=N$
- $-10^9<=v<=10^9$
- $1<=k<=5$

## **Mô tả:**
  Đây là một trong những bài toán sử dụng cây **Segment Tree (Interval Tree)** điển hình, kết hợp kỹ thuật **Lazy Propagation** để tối ưu độ phức tạp của bài toán, trong đó mỗi node trên cây là 1 set các pair (tối đa 5 pair)

  Mỗi truy vấn, nếu x = 1 ta up cây theo đoạn [l, r] một mức k, nếu x = 2 ta in ra set của node **root** trong cây (do node root chứa 5 phần tử lớn nhất)

## **Lưu ý đặc biệt:**
Việc sử dụng **Set** trong **Python** là 1 trong các trở ngại của những người mới bắt đầu chuyển hướng từ **C++** sang **Python** vì:


*   Các phần tử trong set là bất biến, nghĩa là không thể thay đổi giá trị các node trong Python
*   Hàm pop() trong Set là xóa đi 1 phần tử **bất kỳ**, nên không thể xóa giá trị min ở cuối như C++

## **Giải quyết:**


*   Tạo 1 set mới, gán 1 biến lưu set cũ **.pop()**, sau đó có thể thay đổi giá trị của biến đó trước khi lưu vào set mới.
*   Sử dụng hàm **min**, đồng thời có thể tùy biến min bằng cách tạo hàm giá trị trả về mới, sau đó gán nó là key trong hàm min ( VD: `min(t, Key = <hàm giá trị>)` )

## **Tham khảo:**

*   https://vnoi.info/wiki/translate/codeforces/Efficient-and-easy-segment-trees.md
*   https://cp-algorithms.com/data_structures/segment_tree.html

***Mở rộng:***


*   https://www.spoj.com/problems/GSS3/

```python
import sys

def treeAddValue(a, value):
    b = {()}
    b.clear()
    while (len(a) > 0):
        x = a.pop()
        b.add( (x[0] + value, x[1]) )
    return b

def take_first_second(x):
    return (x[1], x[0])

def treeLazy(i):
    lazy[i << 1] += lazy[i]
    lazy[i << 1|1] += lazy[i]

    t[i << 1] = treeAddValue(t[i << 1], lazy[i])
    t[i << 1|1] = treeAddValue(t[i << 1|1], lazy[i])
    lazy[i] = 0

def treeCreate(i, l, r):
    lazy[i] = 0
    if (l == r):
        t[i].clear()
        t[i].add((0, l))
        return
    
    mid = (l + r) // 2
    treeCreate(i << 1, l, mid)
    treeCreate(i << 1|1, mid +1, r)
    t[i] = t[i << 1] | t[i << 1|1]
    while (len(t[i]) > 5):
        t[i].remove(min(t[i]))

def up(i, l, r, u, v, x):
    if (v < l or u > r):
        return
    if (l >= u and r <= v):
        lazy[i] += x
        t[i] = treeAddValue(t[i], x)
        return 
    
    if (lazy[i] != 0):
        treeLazy(i)

    mid = (l+r) // 2
    up(i << 1, l, mid, u, v, x)
    up(i << 1|1, mid+1, r, u, v, x)
    t[i] = t[i << 1] | t[i << 1|1]
    while (len(t[i]) > 5):
        t[i].remove(min(t[i]))

maxn = 50000 * 4 +10
#sys.stdin = open("input1.txt", "r")
n, q = [int(x) for x in input().split()]
lazy = [(0) for i in range(maxn)]
t = [{()} for x in range(maxn)]

treeCreate(1, 1, n)
for i in range(q):
    arr = [int(x) for x in input().split()]
    if arr[0] == 1:
        up(1,1,n,arr[1],arr[2],arr[3])
    else:
        ans = []
        ans = t[1]
        ans = sorted(ans, reverse= True)
        for x in ans:
            print(x[1], end = " ")
            arr[1] -= 1
            if (arr[1] == 0):
                break
        print()
```

# Bài 4: Splitting carrots

## **Tóm tắt đề bài**: 
Cho dãy A gồm N phần tử, cần kiểm tra xem phải bỏ đi bao nhiêu phần tử thì dãy còn lại không thể chia thành 2 phần có tổng bằng nhau

**Ràng buộc:**

- $1<=N<=100$
- $A[i] <= 2000$

## **Mô tả:**
Kỹ thuật sử dụng:
*   QHĐ tổng cơ bản không truy vết
*   Toán học

**Lưu ý**: Ưu tiên kết quả bằng 0 trước khi xét các trường hợp tiếp theo.

## **Tham khảo:**
* https://vnoi.info/wiki/algo/dp/basic-problems.md

```python
n = int(input())
arr = [int(i) for i in input().split()]
sum = 0
chk = [False for i in range(200000+10)]

for x in arr:
    sum += x

if (sum % 2 != 0):
    print("0")
    exit()

sum2 = int(sum / 2)

chk[0] = True
for x in arr:
    if chk[sum2] == True:
        break

    for i in reversed(range(0, sum2 + 1 - x)):
        if chk[i]:
            chk[i + x] = True
            
        if chk[sum2] == True:
            break

if (chk[sum2] == False):
    print("0")
    exit()
else:
    while (True):
        for i in range(n):
            if arr[i] % 2 != 0:
                print("1")
                print(i + 1)
                exit()
            arr[i] = int(arr[i] / 2)
```

# Bài 5: Minesweeper

## **Tóm tắt đề bài**: 
Cho ma trận A kích thước NxM, với a[i][j] là số ô có bom chung cạnh với ô đó ( không tính chính nó). Xuất ra 1 ma trận b tương ứng:

*   b[i][j] = 1 nếu b[i][j] có bom
*   b[i][j] = 0 nếu b[i][j] không có bom

**Ràng buộc :**
- $1<=N,M<=40$
- $0<=a[i][j]<=4$

## **Mô tả:**
Kỹ thuật sử dụng:
*   Đệ quy quay lui (có nhánh cận)

**Lưu ý**: Để giảm độ phức tạp bài toán hết mức có thể, nếu hàm check() trả về True thì kết thúc chương trình đồng nghĩa tìm ra được kết quả bài toán

```python
import sys

def nextTo(i, j):
    s= 0
    if i>0:
        s += a[i-1][j]
    if i<n-1:
        s += a[i+1][j]
    if j>0:
        s += a[i][j-1]
    if j<m-1:
        s += a[i][j+1]
    return s

def check():
    for i in range(1, n):
        for j in range(m):
            #check around b[i-1][j], except a[i][j]
            tmp = nextTo(i-1, j) - a[i][j]
            #check cases of a[i][j]
            a[i][j] = b[i-1][j] - tmp
            if a[i][j] < 0 or a[i][j] > 2:
                return 0
            if nextTo(i-1,j) != b[i-1][j]:
                return 0

    for i in range(n):
        for j in range(m):
            if nextTo(i, j) != b[i][j]:
                return 0
    return 1

def bruteForce(u, v):
    if ck[0]: 
        return
    for i in range(2):
        if ck[0]: 
            return
        a[u][v] = i
        if v > 0 and b[u][v] - a[u][v-1] < 0:
            continue
        if u == 0 and v == m-1:
            ck[0] = check()
            if ck[0]:
                for i in range(n):
                    for j in range(m):
                        print(a[i][j], end = ' ')
                    print()
        else:
            if (u > 0):
                bruteForce(u - 1, v)
            else:
                bruteForce(u, v + 1)

#sys.stdin = open("input1.txt", "r")

n, m = [int(x) for x in input().split()]
a = [[int(0) for j in range(m)] for i in range(n)]
b = [[int(j) for j in input().split()] for i in range(n)]
ck = [0,0]

bruteForce(0,0)
```
